# 1-1 落体の運動

## モデル

### 流体中の運動方程式：
$$ m { dv \over dt } = mg-R-B , v={dx\over dt}$$
ここで，$m$:落体質量,$g$:重力加速度,$R$:流体抵抗,$B$:浮力
$x$は下向きとする

### 係数： 
落体半径$r$,密度$\sigma$の球体とすれば，$mg=(4/3)\pi r^3 \sigma g$，流体の密度を$\rho$とすれば，浮力Bは$B=(4/3)\pi r^3 \rho g$となる．

流体抵抗には，粘性抵抗力，慣性抵抗力があり，それぞれ，速度に比例，速度の2乗に比例する．落体の速度が小さい場合は慣性抵抗を無視できる.
粘性抵抗は速度に比例するため,$-Kv$と書ける．流体の粘性を$\eta$とおくと，ストークスの法則より，$Kv = 6 \pi \eta r$となる．

### 落体の速度変化を求める
密度$\rho$，粘性$\eta$の流体中の半径$r$,密度$\sigma$の球体の落下速度変化を求めるため，運動方程式を解く．

運動方程式の両辺を$m$で割り,$G=(4/3)\pi r^3 (\sigma - \rho) g / m = { (\sigma - \rho)\over \sigma } g $, $k = 6 \pi \eta r / m$とおけば，運動方程式は
$$  { dv \over dt } = G - kv $$
となる．


### オイラー法
方程式を単純差分化し，オイラー法で解く．時間間隔を$\Delta t$,時刻t_iの速度を$v_i$とすると
$$  { dv \over dt } = G - kv  →   { (v_{i+1} - v_i ) \over \Delta t } = G - kv_i $$
のように差分化でき，速度の変化を次式で求めることができる.
$$ v_{i+1} = v_i  + (G - kv_i )\Delta t $$
オイラー法の誤差は$\Delta t^2$オーダーである．

```python
# ----------------------------------------------------------
# 粘性抵抗があるときの落体の運動(オイラー法)
# ----------------------------------------------------------
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# ----------------------------------------------------------
# ---- 問題パラメータ設定  ---
# モデル
g     = 9.8    # 重力加速度 m/s2

sigma = 5000.    # 球体密度 kg/m3
r     = 0.1    # 球体半径 m

rho   = 1000.    # 流体密度 kg/m3
eta   = [1.1, 0.6, 0.3, 0.]  # 流体粘性(-)

# 計算条件
v0    = 0.    # 初期速度 m/s
t0    = 0.    # 初期時間(s)
tend  = 100.   # 終了時間(s)
dt    = 0.01  # タイムステップ幅(時間間隔)

# ----------------------------------------------------------
# --- 定数計算
nstep = int((tend - t0 ) / dt) # 計算ステップ数
m = ( 4./3.) * np.pi * r**3 *  sigma 
G =  ( sigma - rho ) * g / sigma

for iVis in range(0,len(eta)): # 粘性値ループ
    # --- 定数計算
    k = 6. * np.pi * eta[iVis] * r / m
    # --- 初期化
    time = np.zeros(nstep)
    vel  = np.zeros(nstep)
    t = t0
    v = v0
    time[0] = t
    vel[0]  = v
    for i in np.arange(1, nstep): # タイムステップループ
        # 次ステップの計算
        v_new = v + ( G - k * v ) * dt
        # 記録
        time[i] = t
        vel[i]  = v_new
        t_new   = t + dt
        # 記録
        time[i] = t_new
        vel[i]  = v_new
        # 更新
        t = t_new
        v = v_new
        
    # 計算結果をプロット
    plt.plot(time,vel,label = str(eta[iVis]) )
    
    # 終端速度のラインをプロット
    if ( k > 0. ) :
        vt = G / k # 理論上の終端速度(terminal velocity) 
        vel = vt * np.ones(nstep)
        plt.plot(time,vel,'--k')
        
# プロット修飾
plt.legend() 
plt.title("Velocity in Viscous Floyd")
plt.xlabel("time(s)")
plt.ylabel("velocity(m/s)")
```

### ルンゲ・クッタ法

運動方程式を高精度の解法であるルンゲクッタ法により解く.
$F(v) = G - kv$ としたとき, 次ステップの速度を求める式は,

$$ v_{i+1}=v_i + (k_1 + 2k_2 + 2k_3 + k_4) / 6 $$
$$,k_1= F(v_i)\Delta t$$
$$,k_2= F(v_i + k1 / 2.)\Delta t$$
$$,k_3= F(v_i + k2 / 2.)\Delta t$$
$$,k_4= F(v_i + k3)\Delta t$$

となる．ルンゲ・クッタ法の誤差は$\Delta t^5$オーダーである．

```python
# ----------------------------------------------------------
# 粘性抵抗があるときの落体の運動(ルンゲクッタ法)
# ----------------------------------------------------------
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
# ----------------------------------------------------------
# ---- 問題パラメータ設定  ---
# モデル
g     = 9.8    # 重力加速度 m/s2

sigma = 5000.    # 球体密度 kg/m3
r     = 0.1    # 球体半径 m

rho   = 1000.    # 流体密度 kg/m3
eta   = [1.1, 0.6, 0.3, 0.]  # 流体粘性(-)

# 計算条件
v0    = 0.    # 初期速度 m/s
t0    = 0.    # 初期時間(s)
tend  = 100.   # 終了時間(s)
dt    = 0.05  # タイムステップ幅(時間間隔)

# --- 定数計算
nstep = int((tend - t0 ) / dt) # 計算ステップ数
m = ( 4./3.) * np.pi * r**3 *  sigma 
G =  ( sigma - rho ) * g / sigma
# ----------------------------------------------------------
def func(V):
    return G - V * k
# ----------------------------------------------------------
for iVis in range(0,len(eta)): # 粘性値ループ
    # --- 定数計算
    k = 6. * np.pi * eta[iVis] * r / m
    # --- 初期化
    time = np.zeros(nstep)
    vel  = np.zeros(nstep)
    t = t0
    v = v0
    time[0] = t
    vel[0]  = v
    for i in np.arange(1, nstep): # タイムステップループ
        # 次ステップの計算
        k1 = func( v           ) * dt
        k2 = func( v + k1 / 2. ) * dt
        k3 = func( v + k2 / 2. ) * dt
        k4 = func( v + k3      ) * dt
        v_new = v + ( k1 + 2* k2 + 2*k3 + k4 ) / 6.
        t_new = t + dt
        # 記録
        time[i] = t_new
        vel[i]  = v_new
        # 更新
        t = t_new
        v = v_new
        
    # 計算結果をプロット
    plt.plot(time,vel,label = str(eta[iVis]) )
    
    # 終端速度のラインをプロット
    if ( k > 0. ) :
        vt = G / k # 理論上の終端速度(terminal velocity) 
        vel = vt * np.ones(nstep)
        plt.plot(time,vel,'--k')
        
# プロット修飾
plt.legend() 
plt.title("Velocity in Viscous Floyd")
plt.xlabel("time(s)")
plt.ylabel("velocity(m/s)")
```

### scipyライブラリの利用
scipyライブラリの常微分方程式ソルバーodeintを利用して解く


```python
# ----------------------------------------------------------
# 粘性抵抗があるときの落体の運動(odeintを利用)
# ----------------------------------------------------------
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

# ----------------------------------------------------------
# 方程式の積分カーネル関数
# ----------------------------------------------------------
def func(V, t, G, k):
    return G - V * k

# ----------------------------------------------------------
# 問題パラメータ設定  
# ----------------------------------------------------------
# --- モデル ---
g     = 9.8    # 重力加速度 m/s2

sigma = 5000.  # 球体密度 kg/m3
r     = 0.1    # 球体半径 m

rho   = 1000.    # 流体密度 kg/m3
eta   = [1.1, 0.6, 0.3, 0.]  # 流体粘性(-)

# --- 計算条件 ---
v0    = 0.    # 初期速度 m/s
t0    = 0.    # 初期時間(s)
tend  = 100.   # 終了時間(s)
dt    = 0.05  # タイムステップ幅(時間間隔)

# --- 定数計算 ---
nstep = int((tend - t0 ) / dt) # 計算ステップ数
m = ( 4./3.) * np.pi * r**3 *  sigma 
G =  ( sigma - rho ) * g / sigma

# ----------------------------------------------------------
# 設定した粘性ごとに解く  
# ----------------------------------------------------------
for iVis in range(0,len(eta)): # 粘性値ループ
    # --- 定数計算 ---
    k = 6. * np.pi * eta[iVis] * r / m
    
    # --- 初期化 ---
    time = np.arange(t0, tend, dt)
    vel = odeint(func, v0, time, args=(G, k))
        
    # 計算結果をプロット
    plt.plot(time,vel,label = str(eta[iVis]) )
    
    # 終端速度のラインをプロット
    if ( k > 0. ) :
        vt = G / k # 理論上の終端速度(terminal velocity) 
        vel = vt * np.ones(nstep)
        plt.plot(time,vel,'--k')
        
# プロット修飾
plt.legend() 
plt.title("Velocity in Viscous Floyd")
plt.xlabel("time(s)")
plt.ylabel("velocity(m/s)")
```

```python
# ----------------------------------------------------------
# 粘性抵抗があるときの落体の運動(odeintを利用)
#   粘性をスライダーバーでインタラクティブに操作
# ----------------------------------------------------------
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 上微分方程式ソルバー odeint を使う
from scipy.integrate import odeint

# interactを使う
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets

# ----------------------------------------------------------
# 問題パラメータ設定  
# ----------------------------------------------------------
# --- モデル ---
g     = 9.8    # 重力加速度 m/s2

sigma = 5000.  # 球体密度 kg/m3
r     = 0.1    # 球体半径 m

rho   = 1000.    # 流体密度 kg/m3
eta   = [1.1, 0.6, 0.3, 0.]  # 流体粘性(-)

# --- 計算条件 ---
v0    = 0.    # 初期速度 m/s
t0    = 0.    # 初期時間(s)
tend  = 100.   # 終了時間(s)
dt    = 0.05  # タイムステップ幅(時間間隔)

# --- 定数計算 ---
nstep = int((tend - t0 ) / dt) # 計算ステップ数
m = ( 4./3.) * np.pi * r**3 *  sigma 
G =  ( sigma - rho ) * g / sigma

# ----------------------------------------------------------
# 方程式の積分カーネル関数
# ----------------------------------------------------------
def func(V, t, G, k):
    return G - V * k

# ----------------------------------------------------------
# 方程式を解いてグラフを書く
# ----------------------------------------------------------
def solv( eta ):
    # --- 定数計算 ---
    k = 6. * np.pi * eta * r / m
    time = np.arange(t0, tend, dt)

    # --- 解く ---
    vel = odeint(func, v0, time, args=(G, k))
        
    # 計算結果をプロット
    plt.plot(time,vel,label = str(eta) )
    
    # プロットの修飾
    plt.ylim([0,800])
    plt.legend() 
    plt.legend() 
    plt.title("Velocity in Viscous Floyd")
    plt.xlabel("time(s)")
    plt.ylabel("velocity(m/s)")
    plt.title("Velocity in Viscous Floyd")
    plt.xlabel("time(s)")
    plt.ylabel("velocity(m/s)")
    
    # 終端速度のラインをプロット
    if ( k > 0. ) :
        vt = G / k # 理論上の終端速度(terminal velocity) 
        vel = vt * np.ones(nstep)
        plt.plot(time,vel,'--k')

# ----------------------------------------------------------
# スライダーバーでインタラクティブに粘性を設定し，対応する解のグラフを表示
# ----------------------------------------------------------
interact(solv, eta = ( 0.0, 2.0, 0.01 ) ) # widgets.IntSliderWidget( min=0.0, max=2.0, step=0.1, value=1.0) );
```
